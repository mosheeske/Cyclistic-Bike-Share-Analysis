#### Observación: Se eliminaron viajes con duración superior a 2 horas para reducir el impacto de valores atípicos extremos que podrían corresponder a errores de registro o usos no representativos.

```python
# Duración promedio por tipo de usuario

df_total.groupby('member_casual')['ride_length'].mean()
```

```python
# Uso por día vs tipo de usuario

df_total.groupby(['member_casual', 'day_of_week']).size().unstack()
```

```python
# Tipo de bicicleta vs usuario

df_total.groupby(['member_casual', 'rideable_type']).size().unstack()
```

```python
# Duración promedio por tipo de usuario

df_avg = df_total.groupby('member_casual')['ride_length'].mean().reset_index()

# Convertir a minutos para graficar
df_avg['ride_length_min'] = df_avg['ride_length'].dt.total_seconds() / 60

import matplotlib.pyplot as plt

plt.figure()
plt.bar(df_avg['member_casual'], df_avg['ride_length_min'])
plt.title('Duración promedio del viaje por tipo de usuario')
plt.xlabel('Tipo de usuario')
plt.ylabel('Minutos')
plt.show()
```

```python
# Uso por día de la semana

df_days = df_total.groupby(['member_casual', 'day_of_week']).size().unstack()

df_days.T.plot(kind='bar')

plt.title('Número de viajes por día de la semana')
plt.xlabel('Día de la semana (0=Lunes)')
plt.ylabel('Cantidad de viajes')
plt.show()
```

```python
# Tipo de bicicleta por usuario

df_bikes = df_total.groupby(['member_casual', 'rideable_type']).size().unstack()

df_bikes.plot(kind='bar')

plt.title('Tipo de bicicleta por usuario')
plt.xlabel('Tipo de usuario')
plt.ylabel('Cantidad de viajes')
plt.show()
```

```python
# Distribución de duración

df_total['ride_length_min'] = df_total['ride_length'].dt.total_seconds() / 60

import seaborn as sns

plt.figure()
sns.boxplot(data=df_total, x='member_casual', y='ride_length_min')

plt.title('Distribución de duración de viajes por tipo de usuario')
plt.xlabel('Tipo de usuario')
plt.ylabel('Minutos')
plt.show()
```

## Conclusiones

1. ⏱️ Diferencia en duración (INSIGHT PRINCIPAL)

👉 Casual: ~16–17 min
👉 Member: ~11–12 min

💥 Conclusión:

Los usuarios casuales realizan viajes significativamente más largos que los miembros, lo que sugiere un uso principalmente recreativo, mientras que los miembros utilizan el servicio de forma más eficiente y orientada a transporte diario.

2. 📅 Comportamiento por día de la semana

👉 Members:

Uso alto y constante entre semana

👉 Casual:

Pico fuerte en fin de semana (viernes–domingo)

💥 Conclusión:

Los miembros presentan un patrón de uso consistente durante la semana, asociado a desplazamientos laborales, mientras que los usuarios casuales muestran mayor actividad durante fines de semana, indicando un uso recreativo.

3. 🚲 Tipo de bicicleta

👉 Ambos grupos prefieren eléctricas
👉 Pero members tienen mayor volumen total

💥 Conclusión:

Las bicicletas eléctricas dominan en ambos segmentos, lo que indica una preferencia general por comodidad y eficiencia en el desplazamiento.

4. 📦 Distribución (boxplot)

👉 Casual:

Mayor dispersión
Más viajes largos

👉 Member:

Más concentrado

💥 Conclusión:

Los usuarios casuales presentan mayor variabilidad en la duración de sus viajes, reforzando la hipótesis de uso recreativo y no estructurado.

## Recomendaciones

1. Convertir usuarios casuales en miembros

Diseñar campañas dirigidas a usuarios de fin de semana incentivando membresías con beneficios para uso recurrente.

Ejemplo:

descuentos lunes–viernes
trial de membresía


2. Promover uso entre semana en casuales

Crear incentivos para que usuarios casuales utilicen el servicio durante días laborales.

Ejemplo:

promociones “primer viaje laboral gratis”
paquetes semanales


3. Potenciar bicicletas eléctricas

Enfocar estrategias comerciales en bicicletas eléctricas, dado su alto nivel de adopción.

Ejemplo:

membresías con acceso prioritario
campañas de comodidad / rapidez

#### El análisis evidenció diferencias claras en el comportamiento entre usuarios casuales y miembros. Mientras los miembros utilizan el servicio como medio de transporte funcional, los usuarios casuales lo emplean principalmente para actividades recreativas. Estas diferencias representan una oportunidad estratégica para diseñar campañas enfocadas en convertir usuarios casuales en miembros, aumentando así la rentabilidad del negocio.
